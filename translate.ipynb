```python
import pandas as pd
from googletrans import Translator

# Read the CSV file
#tweets = pd.read_csv('input_file.csv')

# Define a function to translate text using Google Translate API
def translate_text(text):
    try:
        translator = Translator()
        return translator.translate(text).text
    except:
        return ""

# Split the dataframe into chunks of 200 rows
tweet_chunks = [tweets[i:i+200] for i in range(0, tweets.shape[0], 200)]

# Loop through each chunk and translate the 'Tweet' column
translated_chunks = []
for chunk in tweet_chunks:
    chunk['Translated_Tweet'] = chunk['Tweet'].apply(translate_text)
    translated_chunks.append(chunk)

# Concatenate the translated chunks back into a single dataframe
translated_tweets = pd.concat(translated_chunks)

# Write the translated tweets to a new CSV file
translated_tweets.to_csv('translated_tweets.csv', index=False)
```

```python
import pandas as pd
import googletrans
import re

# load the csv file
#df = pd.read_csv('your_csv_file.csv')

# create a translator object
translator = googletrans.Translator()

# define the regex pattern to match Hinglish characters
pattern = re.compile("[\u0900-\u097F]+")

# define a function to translate Hinglish to English
def translate_tweet(tweet):
    # translate the tweet using Google Translate
    translation = translator.translate(tweet, src='hi', dest='en')
    # check if the translation is None
    if translation is not None:
        # apply the regex pattern to the translated text
        english_tweet = pattern.sub('', translation.text)
        return english_tweet
    else:
        return ""

# apply the translation function to the 'tweet' column
tyy['Tweet_new'] = tyy['Tweet'].apply(translate_tweet)

# save the translated dataframe to a new csv file
#df.to_csv('translated_file.csv', index=False)
```

```python
import pandas as pd

# define positive, negative, and neutral words
positive_words = ['good', 'great', 'excellent', 'awesome', 'fantastic', 'nice', 'amazing', 'love', 'happy', 'enjoy']
negative_words = ['bad', 'terrible', 'horrible', 'awful', 'dislike', 'unhappy', 'hate', 'disappoint', 'annoy', 'disgust']
neutral_words = ['ok', 'fine', 'average', 'normal', 'typical', 'usual', 'standard', 'regular', 'common', 'routine']

# define function to calculate sentiment
def get_sentiment(tweet):
    positive_count = sum([1 for word in tweet.lower().split() if word in positive_words])
    negative_count = sum([1 for word in tweet.lower().split() if word in negative_words])
    neutral_count = sum([1 for word in tweet.lower().split() if word in neutral_words])
    
    if positive_count > negative_count and positive_count > neutral_count:
        return 'positive'
    elif negative_count > positive_count and negative_count > neutral_count:
        return 'negative'
    else:
        return 'neutral'
    
# add sentiment column to tweets dataframe
tweets['sentiment'] = tweets['Tweet'].apply(get_sentiment)

# display tweets with sentiment
print(tweets[['Tweet', 'sentiment']])
```
