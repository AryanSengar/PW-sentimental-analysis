```python
#import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def cleanTWT(tweet):
    # Remove mentions
    tweet = re.sub(r'@[A-Za-z0-9]+','',tweet)
    
    # Remove hashtags
    tweet = re.sub(r'#', '', tweet)
    
    # Remove retweet indicator
    tweet = re.sub(r'RT[\s]+', '', tweet)
    
    # Remove URLs
    tweet = re.sub(r'https?:\/\/\S+', '', tweet)
    
    # Remove emojis
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    tweet = re.sub(emoj, '', tweet)
    
    # Convert to lowercase
    tweet = tweet.lower()
    
    # Remove punctuation
    tweet = re.sub(r'[^\w\s]','',tweet)
    
    # Tokenize words
    words = tweet.split()
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    # Join words back into a sentence
    clean_tweet = ' '.join(words)
    
    return clean_tweet
```

```python
import pandas as pd
from google.cloud import translate_v2 as translate

# Set up the translation client
translate_client = translate.Client()

# Function to translate text
def translate_text(text, target_language):
    result = translate_client.translate(text, target_language=target_language)
    return result["translatedText"]

# Read in the csv file into a pandas dataframe
#tweets = pd.read_csv("your_file.csv")

# Create an empty list to store the translated text
translated_tweets = []

# Loop through each tweet and translate it using the function
for tweet in tweets["Tweet"]:
    # Skip empty tweets
    if not tweet:
        translated_tweets.append("")
    else:
        try:
            translated_tweet = translate_text(tweet, "en")
            translated_tweets.append(translated_tweet)
        except:
            translated_tweets.append("")
        
# Add the translated tweets to the dataframe as a new column
tweets["Translated_Tweet"] = translated_tweets

# Print the first 5 rows of the dataframe to check the translation
print(tweets.head())
```
